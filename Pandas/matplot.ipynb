```python
import seaborn as sms
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dresses.csv')
```

```python
sms.histplot(y='Sale Price (â‚¹)', x='Rating', data = df, hue= 'Company_name', legend=False, palette= 'YlGnBu')
```

```python
df_sale = df.groupby('Company_name')['Sale Price (â‚¹)'].max().to_list()
df_rating = df.groupby('Company_name')['Rating'].max().to_list()

plt.scatter(df_rating, df_sale, marker='v', color = 'g', alpha= 0.5)
plt.xticks(np.arange(2,5.1,0.5))
plt.yticks(np.arange(0, 1500, 200))
plt.xlabel('Ratings')
plt.ylabel('Sales')
```
