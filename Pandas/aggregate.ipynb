```python
import pandas as pd
```

```python
df = pd.read_csv('dresses.csv')
```

```python
df.head()
```

```python
df['Company_name'].value_counts()
```

```python
df[df['Sale Price (₹)'] > 500]['Company_name'].value_counts()
```

```python
#To get the max sale price by each company
df_new = df.groupby('Company_name')['Sale Price (₹)'].max()
df_new
```

```python
df.groupby('Company name').count()
```

```python
#To get the mean sale price by each company
df.groupby('Company name')['Sale Price (₹)'].mean()
```

```python
#To get the min sale price by each company
df.groupby('Company name')['Sale Price (₹)'].min()
```
