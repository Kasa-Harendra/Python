```python
import pandas as pd
```

```python
df = pd.read_csv('dresses.csv')
```

* **Usage of head() and tail() is already seen**

> **sample( n )** can be used to return n random rows from the df 

```python
df.sample(3)
```

> **DataFrame.loc[rows,cols]** is used to access required rows and columns

Here rows and cols can be an iterable sequence like list, tuple, range()

```python
#Accessing 1st row
df.loc[0]
```

```python
#Accessing 3 rows
df.loc[[0,1,5]]
```

```python
#Accesing rows using slicing
df.loc[5:7]  #Here slicing is inclusive i.e. even the upper bound is considered
```

```python
#Accessing rows and columns
df.loc[[1,2,5],"Company name"]
```

```python
#Accessing multiple rows and cols
df.loc[(5,6,7), ['Company name', 'Sale Price (₹)']]
```

```python
#All rows but only some cols
df.loc[:, ['Company name', 'Sale Price (₹)']]
```

> **DataFrame.iloc(rows, cols)** works with indexes of the columns

```python
#Accessing using iloc()
df.iloc[2:4, [0,3]]  #upper index is not inclusive when working with iloc()
```

```python
df.iloc[2:4, 1:3]
```

```python
max(df["Sale Price (₹)"])
```

```python
df.at[3, 'Company name']
```

```python
df.sort_values("Company name")
```

```python
df.sort_values("Sale Price (₹)", ascending = True)
```

```python
df.sort_values(['Sale Price (₹)', 'Company name'], ascending = [0,1])
```

```python
df[(df['Company name'] == "KOTTY") & (df['Sale Price (₹)'] > 500)]
```

```python
#Returns the records where company names are in provided iterable
df[df['Company name'].isin(['DEKLOOK', 'KOTTY', 'FLOVITTA'])] 
```

```python
df[df['Company name'].str.contains('KO|TT')] #Checks for company names where TT or KO occured
```

```python
#To get number of null values in a particular column
df['Rating'].isna().sum()
```

```python
#To show the null valued rows 
df[df['Rating'].isna()]
```

```python
#T fill the null fields with a particular value
df['Rating'] = df['Rating'].fillna(2.5)
```
